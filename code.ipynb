آمار استنباطی

آیا رابطه ای بین قیمت محصولات با تعداد خریدشان وجود دارد؟

H0: رابطه ای بین قیمت محصولات با تعداد فروش وجود ندارد

H1: رابطه معنا داری بین قیمت محصولات با تعداد فروش وجود دارد

```python
from scipy import stats
shapiro_price = stats.shapiro(df_clean4['selling_price'])
shapiro_sales = stats.shapiro(df_clean4['suggestion_count'])
print(f"آزمون شاپیرو برای قیمت: {shapiro_price}")
print(f"آزمون شاپیرو برای تعداد فروش: {shapiro_sales}")
```

توزیع نرمال نیست پی از تست همبستگی اسپیرمن استفاده میشود

همبستگی اسپیرمن (Spearman)

```python
from scipy.stats import spearmanr

correlation, p_value = spearmanr(df_clean4['selling_price'], df_clean4['suggestion_count'])
print(f"Spearman correlation: {correlation}, p-value: {p_value}")
```

p-value = 3.16e-272:

H0 رد میشود

این مقدار خیلی کوچکتر از 0.05 است، بنابراین رابطه‌ی بین قیمت و تعداد فروش از نظر آماری معنادار است.

توجه به اینکه همبستگی منفی و معنادار است، می‌توان نتیجه گرفت که در این داده‌ها، با افزایش قیمت، تعداد فروش کاهش پیدا می‌کند. این می‌تواند نشان‌دهنده حساسیت بازار به قیمت باشد.

پیشنهاد:

بازار هدف: اگر می‌خواهید محصولاتی با قیمت بالاتر بفروشید، شاید نیاز باشد که در تبلیغات یا استراتژی‌های فروش، مزایای منحصر به فرد محصول را بیشتر نمایش دهید.

قیمت‌گذاری: قیمت پایین‌تر به طور طبیعی ممکن است فروش بالاتری داشته باشد، بنابراین بهتر است استراتژی‌های قیمت‌گذاری رقابتی در نظر گرفته شود.
